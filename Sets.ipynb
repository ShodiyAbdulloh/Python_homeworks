```python
# Set qandey tuzamiz
my_set={20,30,90,50,60}
my_set
```

```python
#bir xil elementni olb tashedi

my_set={20,30,90,50,60,90,30}
my_set
```

```python
# setimizda bor yoki yoq elementni tekshirish
40 in my_set
```

```python
# setimizda bor yoki yoq elementni tekshirish

20 in my_set
```

```python
# elementni qandey qoshamiz

my_set.add("doppi")
my_set
```

```python
#elementni olb tashash

my_set.remove('doppi')
my_set
```

```python
#setni uzunligini olchash

len(my_set)
```

```python
#setimizni ochirib tashlash
my_set.clear()
```

```python
a={0,9,8,7,6}
b={8,7,6,3,2}
a.union(b) # birlashtirish bir xil element bolsa olb tashlaydi
```

```python
a={0,9,8,7,6}
b={8,7,6,3,2}
a.intersection(b) #Ikki setning umumiy bir xil elementlarini olish.
```

```python
a={0,9,8,7,6}
b={8,7,6,3,2}
a.difference(b)                # a setni b setdagi farqi
```

```python
a={0,9,8,7,6}
b={8,7,6,3,2}
a.symmetric_difference(b)  # a va b setning umumiy ozaro farqi 
```

```python
#subset
a={0,9,8,7,6}
b={8,7,0}
b.issubset(a)    # bu yerda b setimiz a setimizdi subseti boladi
```

```python
a={0,9,8,7,6}
b={8,7,0}
a.issuperset(b) # b setimiz a ning superseti boladi
```

```python
a={0,9,8,7,6}
b={8,7,0}
b.issuperset(a)
```
